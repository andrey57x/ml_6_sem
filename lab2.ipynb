```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Я буду использовать готовый датасет Titanic

```python
df = sns.load_dataset('titanic')

df.head()
```

```python
df.isnull().sum()
```

Колонку deck можно просто удалить, так как там слишком много пропусков. Строки с embarked тоже можно удалить, так как там слишком мало пропусков.

```python
clear = df.drop(columns=['deck']).dropna(subset=['embark_town', 'embarked'])
clear.isnull().sum()
```

В колонке age мы заменил возраст на медианный.

```python
clear['age'] = clear['age'].fillna(clear['age'].median())
clear.isnull().sum()
```

От пропусков избавились. Переходим к кодированию категориальных признаков.

```python
clear.info()
```

Закодируем пол и одиночество как 0 и 1.

```python
clear['sex'] = clear['sex'].map({'male': 0, 'female': 1})
clear['alone'] = clear['alone'].map({False: 0, True: 1})
clear.info()
```

Колонки alive, who, embark_town и adult_male можно выкинуть, так как они дублируют уже существующие признаки.

```python
clear.drop(columns=['alive','who','adult_male', 'embark_town'], inplace=True)
clear.info()
```

Класс мы закодируем от 1 до 3, так как его можно упорядочить.

```python
clear['class'] = clear['class'].map({'First': 1, 'Second': 2, 'Third': 3}).astype(int)
clear.info()
```

Закодируем one-hot колонку embarked

```python
embarked_dummies = pd.get_dummies(clear['embarked'], prefix='embarked')
embarked_dummies = embarked_dummies.astype(int)
clear = pd.concat([clear, embarked_dummies], axis=1)
clear = clear.drop('embarked', axis=1)
clear.info()
```

Масштабируем данные.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

clear_scaled = pd.DataFrame(scaler.fit_transform(clear.drop('survived', axis=1)))

clear_scaled.describe()
```
