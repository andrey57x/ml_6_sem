```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_wine

wine = load_wine()

df = pd.DataFrame(wine.data, columns=wine.feature_names)
df['target'] = wine.target
```

В качестве набора данных для лабораторной работы выбран **Wine Recognition Dataset** из библиотеки scikit-learn. 

Данный датасет содержит результаты химического анализа вин, выращенных в одном и том же регионе Италии, но произведенных из трех разных сортов винограда. Задача, для которой обычно используется этот датасет — классификация сорта вина на основе его химических характеристик.

Из полученных данных видно, что датасет состоит из 178 записей и 14 столбцов (включая целевой признак). Все признаки имеют числовой тип данных (float64, а целевая переменная — int32). Как видно из вывода df.info(), пропущенные значения отсутствуют.

Строим графики

```python
sns.set_theme(style="whitegrid")
```

```python
df.hist(figsize=(15, 12), bins=20, color='skyblue', edgecolor='black')
plt.subplots_adjust(hspace=0.5)
plt.show()
```

Гистограммы показывают распределение каждого признака. Некоторое из признаков имеют распределение, близкое к нормальному.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='target', y='alcohol', data=df, palette='Set2')
plt.title('Распределение алкоголя в зависимости от сорта вина')
plt.xlabel('Сорт вина (Target)')
plt.ylabel('Содержание алкоголя')
plt.show()
```

На графике видно, что вино сорта '0' в среднем крепче, чем вино сорта '2', а оно крепче '1'.

```python
selected_features = ['alcohol', 'malic_acid', 'color_intensity', 'flavanoids', 'target']
sns.pairplot(df[selected_features], hue='target', palette='bright')
plt.show()
```

Вычислим корреляцию

```python
plt.figure(figsize=(12, 10))
corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Матрица корреляции признаков')
plt.show()
```

Как мы видим, сильно коррелирует с target flavanoids, od280/0d315 признак и total phenols, а практические не коррелирует ash.
